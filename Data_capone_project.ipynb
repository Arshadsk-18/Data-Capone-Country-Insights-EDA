```python
# which country has the highest ppopulation 
df[df["population"] == df["population"].max()]["country"]
```

```python
# What is the capital of highest populated country
df[df["population"] == df["population"].max() ]["capital_city"]
```

```python
# Which is the least populated country
df[df["population"] == df["population"].min()]['country']
```

```python
# capital of country with least populaton
df[df["population"] == df["population"].min()]['capital_city']
```

```python
# Top 5 countries with highest democratic rate 
df.sort_values("democracy_score",ascending=False).head(5)
```

```python
# Which countries lies in the Eastern Europe region
df[df['region'] == 'Eastern Europe']['country']
```

```python
# Who is the political leader of second highest populated country
df[df["population"]== df["population"].nlargest(2).iloc[1]]['political_leader']
```

```python
# How many political countries with political leader is unknown 
df[df["political_leader"].isna()]['country']
```

```python
# how many countries have REPUBLIC in thier name 
count = 0
def counting_countries(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt

df["country_long"].apply(counting_countries)
print(count)
```

```python
# Which country in African region has the highest population
africa_df = df[df['continent'] == 'Africa']
africa_df[africa_df["population"] == africa_df["population"].max()]["country"]
```

Identified the highest and least populated countries and their capitals.

Extracted the top 5 countries with the highest democracy rate.

Filtered countries belonging to the Eastern Europe region.

Found the political leader of the second most populated country and counted countries with unknown leaders.

Performed text analysis to count how many countries have “Republic” in their name.

Identified the most populated country in the African region.
